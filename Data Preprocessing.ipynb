```python
import pandas as pd
import urllib
import geopandas as gpd
```

```python
crime_url = 'https://data.cityofnewyork.us/api/views/5uac-w243/rows.csv?accessType=DOWNLOAD'
urllib.request.urlretrieve(crime_url, "crimes.csv")
```

```python
crimes_df = pd.read_csv('crimes.csv')
```

```python
crimes_df.head()
```

```python
#crimes_df['crime_date'] = crimes_df['CMPLNT_FR_DT'] + crimes_df['CMPLNT_FR_TM']

#crimes_df_date['crime_date'] = pd.to_datetime(crimes_df['CMPLNT_FR_DT'])

#starting year = 2006:
crimes_df['yearoc'] = crimes_df['CMPLNT_FR_DT'].str[-4:]
```

```python
crimes_df.tail()
```

```python
crimes_df['yearoc'] = crimes_df['yearoc'].astype(int)

crimes_post2006 = crimes_df[crimes_df['yearoc'] >= 2006]
```

```python
# convert the date to datetime
#crimes_post2006['crime_date'] = crimes_post2006['CMPLNT_FR_DT'] + ' ' + crimes_post2006['CMPLNT_FR_TM']

crimes_post2006['crime_date'] = pd.to_datetime(crimes_post2006['CMPLNT_FR_DT'] + ' ' + crimes_post2006['CMPLNT_FR_TM'], infer_datetime_format=True)
```

```python
precinct_shp = 'https://data.cityofnewyork.us/api/geospatial/78dh-3ptz?method=export&format=Shapefile'
urllib.request.urlretrieve(precinct_shp, "precinct_shp.zip")
!unzip precinct_shp.zip
```

```python
#group number of crimes by offense type at precinct level at date and hour:
crimes_post2006['hour'] = crimes_post2006['crime_date'].dt.hour
crimes_post2006['date_crime'] = crimes_post2006['crime_date'].dt.date
```

```python
crimes_post2006.head()
```

```python

crimes_post2006.drop(['X_COORD_CD','Y_COORD_CD','Latitude','Longitude','Lat_Lon','CMPLNT_FR_DT','CMPLNT_FR_TM','CMPLNT_TO_DT','CMPLNT_TO_TM','BORO_NM','STATION_NAME','TRANSIT_DISTRICT','PATROL_BORO','CMPLNT_TO_DT','CMPLNT_TO_TM','HADEVELOPT','HOUSING_PSA','JURIS_DESC','PD_CD','PD_DESC'],axis =1,inplace=True)
```

```python
crimes_post2006.columns
```

```python
crimes_post2006['CRM_ATPT_CPTD_CD'].replace('COMPLETED',0,inplace = True)
crimes_post2006['CRM_ATPT_CPTD_CD'].replace('ATTEMPTED',1,inplace = True)
```

```python
crimes_post2006['LAW_CAT_CD'].replace('FELONY',0,inplace = True)
crimes_post2006['LAW_CAT_CD'].replace('MISDEMEANOR',1,inplace = True)
crimes_post2006['LAW_CAT_CD'].replace('VIOLATION',2,inplace = True)
```

```python
crimes_post2006['LOC_OF_OCCUR_DESC'].replace('INSIDE',0,inplace = True)
crimes_post2006['LOC_OF_OCCUR_DESC'].replace('FRONT OF',1,inplace = True)
crimes_post2006['LOC_OF_OCCUR_DESC'].replace('REAR OF',2,inplace = True)
crimes_post2006['LOC_OF_OCCUR_DESC'].replace('OPPOSITE OF',3,inplace = True)
crimes_post2006['LOC_OF_OCCUR_DESC'].replace('OUTSIDE',4,inplace = True)
```

```python
crimes_post2006.head()
```

```python
crimes_post2006['LOC_OF_OCCUR_DESC'].replace('NaN',-1,inplace = True)
crimes_post2006['LOC_OF_OCCUR_DESC'].fillna(-1,inplace = True)
```

```python
crimes_post2006.loc[crimes_post2006['PARKS_NM'].notnull(), 'PARKS_NM'] = 1
```

```python
crimes_post2006['PARKS_NM'].fillna(0,inplace = True)
```

```python
crimes_post2006.drop(['PREM_TYP_DESC'],axis = 1, inplace = True)
```

```python
crimes_post2006.head()
```

```python
crimes_post2006['SUSP_RACE'].fillna('UNKNOWN',inplace = True)
```

```python
crimes_post2006['SUSP_RACE'].replace('UNKNOWN',0,inplace = True)
crimes_post2006['SUSP_RACE'].replace('BLACK',1,inplace = True)
crimes_post2006['SUSP_RACE'].replace('WHITE HISPANIC',2,inplace = True)
crimes_post2006['SUSP_RACE'].replace('ASIAN/PAC.ISL',3,inplace = True)
crimes_post2006['SUSP_RACE'].replace('WHITE',4,inplace = True)
crimes_post2006['SUSP_RACE'].replace('BLACK HISPANIC',5,inplace = True)
crimes_post2006['SUSP_RACE'].replace('AMER IND',6,inplace = True)
```

```python
crimes_post2006['SUSP_SEX'].fillna('U',inplace = True)
crimes_post2006['SUSP_SEX'].replace('M',1,inplace = True)
crimes_post2006['SUSP_SEX'].replace('F',2,inplace = True)
crimes_post2006['SUSP_SEX'].replace('U',0,inplace = True)
```

```python
crimes_post2006.head()
```

```python
crimes_post2006['VIC_RACE'].replace('UNKNOWN',0,inplace = True)
crimes_post2006['VIC_RACE'].replace('BLACK',1,inplace = True)
crimes_post2006['VIC_RACE'].replace('WHITE HISPANIC',2,inplace = True)
crimes_post2006['VIC_RACE'].replace('ASIAN/PAC.ISL',3,inplace = True)
crimes_post2006['VIC_RACE'].replace('WHITE',4,inplace = True)
crimes_post2006['VIC_RACE'].replace('BLACK HISPANIC',5,inplace = True)
crimes_post2006['VIC_RACE'].replace('AMER IND',6,inplace = True)

crimes_post2006['VIC_SEX'].fillna('U',inplace = True)
crimes_post2006['VIC_SEX'].replace('M',1,inplace = True)
crimes_post2006['VIC_SEX'].replace('F',2,inplace = True)
crimes_post2006['VIC_SEX'].replace('U',0,inplace = True)
```

```python
crimes_post2006['VIC_SEX'].replace('D',0,inplace = True)
crimes_post2006['VIC_SEX'].replace('E',0,inplace = True)
```

```python
crimes_post2006.head()
```

```python
crimes_post2006['VIC_AGE_GROUP'].unique()

ration_age1 = crimes_post2006[(crimes_post2006['VIC_AGE_GROUP'] == '18-24' ) | (crimes_post2006['VIC_AGE_GROUP'] == '25-44' ) | (crimes_post2006['VIC_AGE_GROUP'] == '45-64' ) | (crimes_post2006['VIC_AGE_GROUP'] == '<18' ) | (crimes_post2006['SUSP_AGE_GROUP'] == '65+') | (crimes_post2006['SUSP_AGE_GROUP'] == 'UNKNOWN' )]
ration_age1.groupby(by='VIC_AGE_GROUP').count()
```

```python
irration_age1 = crimes_post2006[~((crimes_post2006['VIC_AGE_GROUP'] == '18-24' ) | (crimes_post2006['VIC_AGE_GROUP'] == '25-44' ) | (crimes_post2006['VIC_AGE_GROUP'] == '45-64' ) | (crimes_post2006['VIC_AGE_GROUP'] == '<18' ) | (crimes_post2006['SUSP_AGE_GROUP'] == '65+')| (crimes_post2006['SUSP_AGE_GROUP'] == '<18' ) | (crimes_post2006['SUSP_AGE_GROUP'] == '65+') | (crimes_post2006['SUSP_AGE_GROUP'] == 'UNKNOWN' ))]
irration_age1['VIC_AGE_GROUP'].str.replace('.*','18-24')

irration_age1['VIC_AGE_GROUP'].fillna('18-24',inplace=True)
```

```python
crimes_post2006['SUSP_AGE_GROUP'].fillna('UNKNOWN',inplace = True)
```

```python
crimes_post2006.VIC_AGE_GROUP.unique()
```

```python
crimes_post2006 = pd.concat([irration_age1, ration_age1],axis=0, ignore_index=True)
crimes_post2006
```

```python
crimes_post2006['SUSP_AGE_GROUP'].unique()

ration_age = crimes_post2006[(crimes_post2006['SUSP_AGE_GROUP'] == '18-24' ) | (crimes_post2006['SUSP_AGE_GROUP'] == '25-44' ) | (crimes_post2006['SUSP_AGE_GROUP'] == '45-64' ) | (crimes_post2006['SUSP_AGE_GROUP'] == '<18' ) | (crimes_post2006['SUSP_AGE_GROUP'] == '65+') | (crimes_post2006['SUSP_AGE_GROUP'] == 'UNKNOWN' )]
ration_age.groupby(by='SUSP_AGE_GROUP').count()
```

```python
irration_age = crimes_post2006[~((crimes_post2006['SUSP_AGE_GROUP'] == '18-24' ) | (crimes_post2006['SUSP_AGE_GROUP'] == '25-44' ) | (crimes_post2006['SUSP_AGE_GROUP'] == '45-64' ) | (crimes_post2006['SUSP_AGE_GROUP'] == '<18' ) | (crimes_post2006['SUSP_AGE_GROUP'] == '65+')| (crimes_post2006['SUSP_AGE_GROUP'] == '<18' ) | (crimes_post2006['SUSP_AGE_GROUP'] == '65+') | (crimes_post2006['SUSP_AGE_GROUP'] == 'UNKNOWN' ))]
irration_age['SUSP_AGE_GROUP'].str.replace('.*','18-24')

irration_age['SUSP_AGE_GROUP'].fillna('18-24',inplace=True)
```

```python
crimes_post2006 = pd.concat([irration_age, ration_age],axis=0, ignore_index=True)
```

```python
crimes_post2006
```

```python
one_hot_CRM_ATPT_CPTD_CD = pd.get_dummies(crimes_post2006['CRM_ATPT_CPTD_CD'], prefix='CRM_ATPT_CPTD_CD')
one_hot_CRM_ATPT_CPTD_CD

one_hot_LAW_CAT_CD = pd.get_dummies(crimes_post2006['LAW_CAT_CD'], prefix='LAW_CAT_CD')
one_hot_LAW_CAT_CD

one_hot_LOC_OF_OCCUR_DESC = pd.get_dummies(crimes_post2006['LOC_OF_OCCUR_DESC'], prefix='LOC_OF_OCCUR_DESC')
one_hot_LOC_OF_OCCUR_DESC

one_hot_SUSP_RACE = pd.get_dummies(crimes_post2006['SUSP_RACE'], prefix='SUSP_RACE')
one_hot_SUSP_RACE

one_hot_SUSP_SEX = pd.get_dummies(crimes_post2006['SUSP_SEX'], prefix='SUSP_SEX')
one_hot_SUSP_SEX

one_hot_SUSP_AGE_GROUP = pd.get_dummies(crimes_post2006['SUSP_AGE_GROUP'], prefix='SUSP_AGE_GROUP')
one_hot_SUSP_AGE_GROUP

one_hot_VIC_AGE_GROUP = pd.get_dummies(crimes_post2006['VIC_AGE_GROUP'], prefix='VIC_AGE_GROUP')
one_hot_VIC_AGE_GROUP

one_hot_VIC_RACE = pd.get_dummies(crimes_post2006['VIC_RACE'], prefix='VIC_RACE')
one_hot_VIC_RACE

one_hot_VIC_SEX = pd.get_dummies(crimes_post2006['VIC_SEX'], prefix='VIC_SEX')
one_hot_VIC_SEX
```

```python
one_hot_hour = pd.get_dummies(crimes_post2006['hour'], prefix='hour')
one_hot_hour
```

```python
DL_DF_NEW =  pd.concat([crimes_post2006, one_hot_CRM_ATPT_CPTD_CD,one_hot_LAW_CAT_CD,
one_hot_LOC_OF_OCCUR_DESC,one_hot_SUSP_RACE,
one_hot_SUSP_SEX,one_hot_SUSP_AGE_GROUP,one_hot_VIC_AGE_GROUP,one_hot_VIC_RACE,one_hot_VIC_SEX,one_hot_hour], axis=1)
```
